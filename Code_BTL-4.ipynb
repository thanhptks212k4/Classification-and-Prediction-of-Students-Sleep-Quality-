## Sử dụng phương pháp Elbow để tìm ra K tối ưu

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Extract relevant columns
data_subset = label_data[['Sleep_Hours', 'Stress_Level']]

# Convert Stress_Level to numerical values using LabelEncoder
label_encoder = LabelEncoder()
data_subset['Stress_Level'] = label_encoder.fit_transform(data_subset['Stress_Level'])

# Normalize the data
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data_subset)

# Elbow method to find the optimal number of clusters
inertia = []
K_range = range(1, 11)  # Test K from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_normalized)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.xticks(K_range)
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Giả sử bạn đã có bộ dữ liệu label_data

# Extract relevant columns
data_subset = label_data[['Sleep_Hours', 'Stress_Level']]

# Convert Stress_Level to numerical values using LabelEncoder
label_encoder = LabelEncoder()
data_subset['Stress_Level'] = label_encoder.fit_transform(data_subset['Stress_Level'])

# Normalize the data
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data_subset)

# Phương pháp Elbow để tìm số cụm tối ưu
inertia_values = []
silhouette_scores = []

# Kiểm tra số cụm từ 2 đến min(7, 10) vì dữ liệu chỉ có 7 mẫu
possible_clusters = range(2, min(8, 11))  # Từ 2 đến 7 cụm (vì bạn có 7 mẫu)
for n_clusters in possible_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_normalized)

    # Inertia để vẽ Elbow curve
    inertia_values.append(kmeans.inertia_)

    # Tính Silhouette Score nếu số cụm > 1 và < số mẫu
    if n_clusters > 1 and n_clusters < len(data_normalized):
        score = silhouette_score(data_normalized, kmeans.labels_)
        silhouette_scores.append(score)
    else:
        silhouette_scores.append(np.nan)  # Nếu không hợp lệ thì thêm giá trị NaN

# Vẽ Elbow curve để tìm số cụm tối ưu
plt.figure(figsize=(8, 6))
plt.plot(possible_clusters, inertia_values, marker='o', linestyle='-', color='b')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()

# Vẽ Silhouette Score để đánh giá độ phân cụm
plt.figure(figsize=(8, 6))
plt.plot(possible_clusters, silhouette_scores, marker='o', linestyle='-', color='r')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()

# Chọn số cụm tối ưu dựa trên Elbow hoặc Silhouette Score
optimal_k = 5  # Thay đổi số cụm nếu cần, có thể chọn theo Elbow method hoặc giá trị Silhouette Score tối ưu

# Áp dụng KMeans với số cụm tối ưu
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data_subset['Cluster'] = kmeans.fit_predict(data_normalized)

# In ra các cụm đã phân loại
print(data_subset.head())

# Tính Silhouette Score cho phân cụm đã thực hiện
sil_score = silhouette_score(data_normalized, data_subset['Cluster'])
print(f"Silhouette Score for k={optimal_k}: {sil_score:.2f}")
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import copy
import pandas as pd
from sklearn.metrics import silhouette_score, davies_bouldin_score

class KMeans:
    def __init__(self, n_cluster=5, random_state=42):  # Thay đổi n_cluster thành 5
        self.n_cluster = n_cluster
        self.random_state = random_state

    def fit(self, label_data):
        self.X = label_data[['Sleep_Hours', 'Stress_Level']]  # Use the first two columns for clustering
        self.m = self.X.shape[0]  # number of training examples
        self.n = self.X.shape[1]  # number of features
        initial_centroids = self.initialize_centroids()
        self.plot_initial_centroids(initial_centroids)
        self.clustering(initial_centroids)

    def initialize_centroids(self):
        initial_centroids = []
        random.seed(self.random_state)

        for i in range(self.n_cluster):
            initial_centroids.append(np.ravel(self.X.iloc[(random.randint(0, self.m - 1)), :]))

        return np.array(initial_centroids)

    def plot_initial_centroids(self, initial_centroids):
        plt.scatter(self.X.iloc[:, 0], self.X.iloc[:, 1], c='#000000', s=7, label='Data Points')
        plt.scatter(initial_centroids[:, 0], initial_centroids[:, 1], marker='*', s=120, c='r', label='Initial Centroids')
        plt.title('Initial Random Cluster Centers')
        plt.xlabel('Sleep Hours')
        plt.ylabel('Stress Level')
        plt.legend()
        plt.show()

    def clustering(self, centroids):
        old_centroids = np.zeros(centroids.shape)
        stopping_criteria = 0.0001
        self.iterating_count = 0

        while self.euclidean_distance(old_centroids, centroids) > stopping_criteria:
            clusters = np.zeros(len(self.X))

            # Assigning each value to its closest cluster
            for i in range(self.m):
                distances = []
                for j in range(len(centroids)):
                    distances.append(self.euclidean_distance(self.X.iloc[i, :], centroids[j]))
                cluster = np.argmin(distances)
                clusters[i] = cluster

            # Storing the old centroid values to compare centroid moves
            old_centroids = copy.deepcopy(centroids)

            # Finding the new centroids
            for i in range(self.n_cluster):
                points = [self.X.iloc[j, :] for j in range(len(self.X)) if clusters[j] == i]
                centroids[i] = np.mean(points, axis=0) if points else centroids[i]

            self.plot_centroids(centroids, clusters)
            self.iterating_count += 1

        self.final_clusters = clusters
        self.centroids = centroids

    def plot_centroids(self, centroids, clusters):
        colors = plt.cm.get_cmap('tab10', self.n_cluster)  # Tạo bảng màu cho số cụm
        fig, ax = plt.subplots()
        for i in range(self.n_cluster):
            points = np.array([self.X.iloc[j, :].values for j in range(len(self.X)) if clusters[j] == i])
            ax.scatter(points[:, 0], points[:, 1], s=7, c=[colors(i)], label=f'Cluster {i + 1}')

        ax.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=120, c='#000000', label='Centroids')

        plt.title('k-Means Clustering')
        plt.xlabel('Sleep Hours')
        plt.ylabel('Stress Level')
        plt.legend()
        plt.show()

    def euclidean_distance(self, a, b):
        return np.sqrt(np.sum((np.array(a) - np.array(b)) ** 2))

# Khởi tạo và chạy mô hình KMeans với k = 5
kmeans = KMeans(n_cluster=5, random_state=42)
kmeans.fit(label_data)

# Gán nhãn cụm vào dataset
label_data['Cluster'] = kmeans.final_clusters.astype(int)


```

```python
import numpy as np
import random
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans as SKMeans
import matplotlib.pyplot as plt

class OptimizedKMeans:
    def __init__(self, random_state=42):
        self.random_state = random_state

    def fit(self, label_data, n_cluster):
        # Chuẩn hóa dữ liệu
        self.X = label_data[['Sleep_Hours', 'Stress_Level']]
        self.scaler = StandardScaler()
        self.X_scaled = self.scaler.fit_transform(self.X)  # Normalize the data

        # Khởi tạo KMeans từ scikit-learn
        kmeans = SKMeans(n_clusters=n_cluster, random_state=self.random_state, n_init=10, init='k-means++')
        kmeans.fit(self.X_scaled)

        # Lưu lại các kết quả
        self.centroids = kmeans.cluster_centers_
        self.labels = kmeans.labels_
        self.inertia = kmeans.inertia_

        # Gán nhãn cụm vào dataset
        label_data['Cluster'] = self.labels

        return label_data

# Giả sử label_data là bộ dữ liệu của bạn
# Chắc chắn rằng bộ label_data đã được định nghĩa ở trước đó

# Thử nhiều giá trị của n_clusters và tính Silhouette Score
silhouette_scores = []
possible_clusters = range(2, 6)  # Từ 2 đến 5 cụm
for n_clusters in possible_clusters:
    kmeans = OptimizedKMeans(random_state=42)
    result = kmeans.fit(label_data, n_clusters)

    # Tính Silhouette Score cho mỗi k
    score = silhouette_score(kmeans.X_scaled, result['Cluster'])
    silhouette_scores.append(score)
    print(f"Silhouette Score for n_clusters={n_clusters}: {score:.2f}")

# Vẽ biểu đồ để tìm số cụm tối ưu
plt.plot(possible_clusters, silhouette_scores, marker='o')
plt.title("Silhouette Score vs. Number of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Định nghĩa thuật toán Naive Bayes với Laplace Smoothing
class NaiveBayes:
    def __init__(self, alpha=1):  # Thêm tham số Laplace Smoothing
        self.alpha = alpha
        self.labels = []
        self.result_value_count = {}
        self.result_values = []
        self.result_label = ""
        self.attribute = {}

    def fit(self, X, Y):
        self.labels = Y[:-1]
        self.result_label = Y[-1]
        self.result_values = X[Y[-1]].unique()

        # Đếm số lần xuất hiện của các giá trị của biến kết quả
        for result_value in self.result_values:
            self.result_value_count[result_value] = (X[Y[-1]] == result_value).sum()

        # Đếm số lần xuất hiện của mỗi giá trị thuộc tính kèm với giá trị của biến kết quả
        for attribute in self.labels:
            attribute_values = X[attribute].unique()

            if attribute not in self.attribute:
                self.attribute[attribute] = {}

            for attribute_value in attribute_values:
                if attribute_value not in self.attribute[attribute]:
                    self.attribute[attribute][attribute_value] = {}

                for result_value in self.result_values:
                    if result_value not in self.attribute[attribute][attribute_value]:
                        self.attribute[attribute][attribute_value][result_value] = 0
                    self.attribute[attribute][attribute_value][result_value] = (
                        (X[attribute] == attribute_value) & (X[Y[-1]] == result_value)
                    ).sum()

    def predict(self, attribute_value):
        result = ""
        max_prob = -1
        for result_value in self.result_values:
            prob = self.result_value_count[result_value] + self.alpha
            prob /= sum(self.result_value_count.values()) + len(self.result_values) * self.alpha
            for key in attribute_value.keys():
                num = self.attribute[key][attribute_value[key]].get(result_value, 0) + self.alpha
                den = self.result_value_count[result_value] + len(self.attribute[key]) * self.alpha
                prob *= num / den
            if prob > max_prob:
                max_prob = prob
                result = result_value
        return result

    def evaluate(self, X):
        correct_count = 0
        for i in range(len(X)):
            row = X.iloc[i]
            attribute_value = {}
            for attribute in self.labels:
                attribute_value[attribute] = row[attribute]
            result_predict = self.predict(attribute_value)
            result_confirm = row[self.result_label]
            if result_predict == result_confirm:
                correct_count += 1
        return correct_count / len(X) * 100
# Sao chép dữ liệu để xử lý
processed_data1 = df.copy()

# Hàm phân loại Sleep_Quality
def categorize_sleep_quality(x):
    if x >= 7:
        return 'Tốt'
    elif 5 <= x < 7:
        return 'Bình thường'
    else:
        return 'Kém'

# Đọc dữ liệu


# Tiền xử lý dữ liệu
processed_data1['Sleep_Quality_Category'] = processed_data1['Sleep_Quality'].apply(categorize_sleep_quality)

# Chuyển đổi các cột số thành dạng phân loại với nhiều bins hơn
processed_data1['Sleep_Hours'] = pd.cut(processed_data1['Sleep_Hours'].astype(float), bins=4, labels=["Very Low", "Low", "Medium", "High"])
processed_data1['Physical_Activity'] = pd.cut(processed_data1['Physical_Activity'].astype(float), bins=4, labels=["Very Low", "Low", "Medium", "High"])
processed_data1['Study_Hours'] = pd.cut(processed_data1['Study_Hours'].astype(float), bins=4, labels=["Very Low", "Low", "Medium", "High"])


# Chọn các cột đặc trưng và mục tiêu
categorical_features = ['Gender', 'Sleep_Hours', 'Physical_Activity', 'Study_Hours','Stress_Level']
target_column = 'Sleep_Quality_Category'
processed_data = processed_data1[categorical_features + [target_column]]

# Tách dữ liệu thành tập huấn luyện và tập kiểm tra
train_df, test_df = train_test_split(processed_data, test_size=0.2, random_state=42)

# Khởi tạo và huấn luyện thuật toán Naive Bayes
nb_algorithm = NaiveBayes(alpha=1)  # Sử dụng Laplace Smoothing
nb_algorithm.fit(train_df, processed_data.columns)

# Đánh giá mô hình trên tập kiểm tra
accuracy = nb_algorithm.evaluate(test_df)
print(f"Độ chính xác của mô hình: {accuracy:.2f}%")

```

```python
# Encode categorical variables and bin Sleep_Quality
from sklearn.preprocessing import LabelEncoder

# Copy the dataset for processing
processed_data = df.copy()

# Encode 'Gender' and 'Stress_Level'
label_encoders = {
    col: LabelEncoder().fit(processed_data[col]) for col in ['Gender', 'Stress_Level']
}
for col, encoder in label_encoders.items():
    processed_data[col] = encoder.transform(processed_data[col])

# Bin 'Sleep_Quality' into categories: Tốt (7+), Bình thường (5-6), Kém (<5)
def categorize_sleep_quality(quality):
    if quality >= 7:
        return 'Tốt'
    elif 5 <= quality < 7:
        return 'Bình thường'
    else:
        return 'Kém'

processed_data['Sleep_Quality_Category'] = processed_data['Sleep_Quality'].apply(categorize_sleep_quality)

import numpy as np

# Gaussian Naive Bayes implementation
class GaussianNaiveBayes:
    def __init__(self):
        self.class_stats = {}

    def fit(self, X, y):
        """Train the model: compute mean and variance for each class."""
        self.classes = np.unique(y)
        for cls in self.classes:
            features = X[y == cls]
            self.class_stats[cls] = {
                "mean": features.mean(axis=0),
                "var": features.var(axis=0),
                "prior": len(features) / len(X)
            }

    def gaussian_pdf(self, x, mean, var):
        """Compute Gaussian Probability Density Function."""
        coef = 1.0 / np.sqrt(2.0 * np.pi * var)
        exponent = np.exp(-((x - mean) ** 2) / (2 * var))
        return coef * exponent

    def predict(self, X):
        """Predict class labels for X."""
        predictions = []
        for x in X:
            posteriors = {}
            for cls, stats in self.class_stats.items():
                prior = np.log(stats["prior"])
                conditional = np.sum(
                    np.log(self.gaussian_pdf(x, stats["mean"], stats["var"]))
                )
                posteriors[cls] = prior + conditional
            predictions.append(max(posteriors, key=posteriors.get))
        return np.array(predictions)

# Prepare data for training and testing
features = ['Sleep_Hours', 'Social_Hours', 'Physical_Activity', 'Extracurricular_Hours', 'Study_Hours']
X = processed_data[features].values
y = processed_data['Sleep_Quality_Category'].values

# Split into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train and test Gaussian Naive Bayes
gnb = GaussianNaiveBayes()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

# Evaluate the model
from sklearn.metrics import classification_report
classification_report(y_test, y_pred)

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder

# Thiết lập các giá trị cv cần thử (từ 2 đến 20)
cv_values = range(2,11)

# Tạo tham số cho KNN
param_grid = {'n_neighbors': range(1, 50)}
# Đọc dữ liệu từ tệp CSV

data = label_data.copy()

# Xác định các cột số và cột phân loại
numerical_features = ['Age', 'Sleep_Hours', 'Social_Hours', 'Physical_Activity', 'Extracurricular_Hours', 'Study_Hours', 'Stress_Level']
categorical_features = ['Gender']


# Kiểm tra và xử lý dữ liệu bị thiếu hoặc không hợp lệ
for col in numerical_features:
    if data[col].dtype == 'object':
        # Chuyển đổi các giá trị không hợp lệ thành NaN
        data[col] = pd.to_numeric(data[col], errors='coerce')
    # Thay thế NaN bằng giá trị trung bình của cột
    data[col].fillna(data[col].mean(), inplace=True)

for col in categorical_features:
    data[col] = data[col].fillna('Unknown')

# Tách dữ liệu thành đầu vào và đầu ra
X = data.drop('Sleep_Quality', axis=1)
y = data['Sleep_Quality']
# Biến đổi dữ liệu
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

X_preprocessed = preprocessor.fit_transform(X)
# Chia dữ liệu thành tập huấn luyện và kiểm thử
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.3, random_state=42)

# Lưu kết quả tốt nhất
best_cv = None
best_k = None
best_score = 0

for cv in cv_values:
    # Tạo và chạy GridSearchCV
    grid_search = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=param_grid, cv=cv, scoring='accuracy', verbose=0)
    grid_search.fit(X_train, y_train)
    
    # Lấy giá trị tốt nhất của k và độ chính xác tốt nhất từ GridSearchCV
    current_best_k = grid_search.best_params_['n_neighbors']
    current_best_score = grid_search.best_score_
    
    # Kiểm tra kết quả
    print(f"CV={cv} => Best k={current_best_k}, Best CV Accuracy={current_best_score:.4f}")

    # Cập nhật nếu tìm thấy giá trị tốt hơn
    if current_best_score > best_score:
        best_score = current_best_score
        best_cv = cv
        best_k = current_best_k

# In ra giá trị tốt nhất sau khi thử nghiệm
print(f"\nOptimal CV: {best_cv}, Best k: {best_k}, Best Accuracy: {best_score:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report, accuracy_score
import tkinter as tk
from tkinter import messagebox

# Đọc dữ liệu từ tệp CSV
data = label_data.copy()

# Xác định các cột số và cột phân loại
numerical_features = ['Age', 'Sleep_Hours', 'Social_Hours', 'Physical_Activity', 'Extracurricular_Hours', 'Study_Hours', 'Stress_Level']
categorical_features = ['Gender']


# Kiểm tra và xử lý dữ liệu bị thiếu hoặc không hợp lệ
for col in numerical_features:
    if data[col].dtype == 'object':
        # Chuyển đổi các giá trị không hợp lệ thành NaN
        data[col] = pd.to_numeric(data[col], errors='coerce')
    # Thay thế NaN bằng giá trị trung bình của cột
    data[col].fillna(data[col].mean(), inplace=True)

for col in categorical_features:
    data[col] = data[col].fillna('Unknown')

# Tách dữ liệu thành đầu vào và đầu ra
X = data.drop('Sleep_Quality', axis=1)
y = data['Sleep_Quality']

# Biến đổi dữ liệu
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

X_preprocessed = preprocessor.fit_transform(X)

# Chia dữ liệu thành tập huấn luyện và kiểm thử
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.3, random_state=42)

# Đảm bảo chỉ số của X_train và X_test là liên tục
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

# Tính khoảng cách Euclidean
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# Cài đặt thuật toán KNN thủ công
class MyKNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = []
        for x in X_test:
            # Tính khoảng cách giữa điểm x và tất cả các điểm trong tập huấn luyện
            distances = [euclidean_distance(x, train_point) for train_point in self.X_train]
            # Sắp xếp các khoảng cách và lấy k láng giềng gần nhất
            k_indices = np.argsort(distances)[:self.k]
            k_nearest_labels = [self.y_train[i] for i in k_indices]
            # Dự đoán nhãn dựa trên láng giềng gần nhất
            most_common = np.bincount(k_nearest_labels).argmax()
            predictions.append(most_common)
        return predictions

# Khởi tạo và huấn luyện mô hình KNN của bạn
model = MyKNN(k=12)
model.fit(X_train, y_train)

# Dự đoán và đánh giá mô hình
y_pred = model.predict(X_test)

# Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN: {100 * accuracy:.2f} %")
print("Classification Report:\n", classification_report(y_test, y_pred))

# Hàm dự đoán chất lượng giấc ngủ từ đầu vào người dùng
def predict_sleep_quality_gui():
    def submit():
        try:
            user_input = {
                'Age': int(age_entry.get()),
                'Sleep_Hours': float(sleep_entry.get()),
                'Social_Hours': float(social_entry.get()),
                'Physical_Activity': float(physical_entry.get()),
                'Extracurricular_Hours': float(extracurricular_entry.get()),
                'Study_Hours': float(study_entry.get()),
                'Stress_Level': int(stress_var.get()),
                'Gender': gender_var.get()
            }

            total_hours = user_input['Sleep_Hours'] + user_input['Social_Hours'] + \
                          user_input['Physical_Activity'] + user_input['Extracurricular_Hours'] + \
                          user_input['Study_Hours']

            if total_hours > 24:
                messagebox.showerror("Error", "Tổng thời gian của các hoạt động không được vượt quá 24 giờ.")
                return

            user_input_df = pd.DataFrame([user_input])
            user_input_preprocessed = preprocessor.transform(user_input_df)
            prediction = model.predict(user_input_preprocessed)
            messagebox.showinfo("Prediction", f"Dự đoán chất lượng giấc ngủ: {prediction[0]}")
        except ValueError as ve:
            messagebox.showerror("Error", f"Nhập liệu không hợp lệ: {ve}")
        except Exception as e:
            messagebox.showerror("Error", f"Có lỗi xảy ra: {e}")

    # Tạo giao diện nhập liệu
    root = tk.Tk()
    root.title("Sleep Quality Prediction")

    tk.Label(root, text="Age:").grid(row=0, column=0)
    age_entry = tk.Entry(root)
    age_entry.grid(row=0, column=1)

    tk.Label(root, text="Sleep Hours:").grid(row=1, column=0)
    sleep_entry = tk.Entry(root)
    sleep_entry.grid(row=1, column=1)

    tk.Label(root, text="Social Hours:").grid(row=2, column=0)
    social_entry = tk.Entry(root)
    social_entry.grid(row=2, column=1)

    tk.Label(root, text="Physical Activity:").grid(row=3, column=0)
    physical_entry = tk.Entry(root)
    physical_entry.grid(row=3, column=1)

    tk.Label(root, text="Extracurricular Hours:").grid(row=4, column=0)
    extracurricular_entry = tk.Entry(root)
    extracurricular_entry.grid(row=4, column=1)

    tk.Label(root, text="Study Hours:").grid(row=5, column=0)
    study_entry = tk.Entry(root)
    study_entry.grid(row=5, column=1)

    tk.Label(root, text="Stress Level:").grid(row=6, column=0)
    stress_var = tk.StringVar(value="1")
    tk.Entry(root, textvariable=stress_var).grid(row=6, column=1)

    tk.Label(root, text="Gender:").grid(row=7, column=0)
    gender_var = tk.StringVar(value="Male")
    tk.OptionMenu(root, gender_var, "Male", "Female", "Other").grid(row=7, column=1)

    submit_button = tk.Button(root, text="Submit", command=submit)
    submit_button.grid(row=8, column=0, columnspan=2)

    root.mainloop()

# Gọi hàm để hiển thị giao diện nhập liệu
predict_sleep_quality_gui()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Hàm kiểm tra độ tối ưu của degree và vẽ biểu đồ
def find_optimal_degree(X, y, max_degree=5):
    results = []

    # Chia dữ liệu
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Chuẩn hóa dữ liệu
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Thử nghiệm với các giá trị degree
    for degree in range(1, max_degree + 1):
        # Tạo Polynomial Features
        poly = PolynomialFeatures(degree=degree)
        X_train_poly = poly.fit_transform(X_train_scaled)
        X_test_poly = poly.transform(X_test_scaled)

        # Huấn luyện mô hình
        model = LinearRegression()
        model.fit(X_train_poly, y_train)

        # Dự đoán trên tập kiểm tra
        y_pred = model.predict(X_test_poly)

        # Tính toán R2 score
        r2 = r2_score(y_test, y_pred)

        # Lưu kết quả
        results.append((degree, r2))

    # Trả về kết quả dưới dạng DataFrame
    return pd.DataFrame(results, columns=["Degree", "R2"])

# Ví dụ sử dụng với dữ liệu
# Giả sử label_data là DataFrame đã được chuẩn bị trước với các đặc trưng và nhãn
X = label_data[feature_columns]
y = label_data["Sleep_Quality"]

# Tính toán R2 cho các bậc đa thức
results = find_optimal_degree(X, y, max_degree=5)
print(results)

# Vẽ biểu đồ R2 theo Degree
plt.figure(figsize=(8, 6))
plt.plot(results["Degree"], results["R2"], marker='o', linestyle='-', color='b', label="R2 Score")
plt.xlabel("Degree of Polynomial")
plt.ylabel("R2 Score")
plt.title("R2 Score vs Degree of Polynomial")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

# Xác định đặc trưng (X) và nhãn (y)
feature_columns = ["Age", "Gender", "Sleep_Hours", "Social_Hours",
                   "Physical_Activity", "Extracurricular_Hours",
                   "Study_Hours", "Stress_Level"]
X = label_data[feature_columns].values  # Chọn các cột đặc trưng
y = label_data["Sleep_Quality"].values  # Cột mục tiêu

# Chuẩn hóa dữ liệu
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X_scaled = (X - mean) / std

# Tạo Polynomial Features (bậc 4)
def polynomial_features(X, degree):
    n_samples, n_features = X.shape
    combinations = [np.ones(n_samples)]
    for d in range(1, degree + 1):
        for items in np.ndindex(*(n_features,) * d):
            combinations.append(np.prod(X[:, items], axis=1))
    return np.vstack(combinations).T

# Thêm cột bias (cột 1s) vào X_poly
X_poly = np.hstack([np.ones((X_poly.shape[0], 1)), X_poly])

# Chia dữ liệu thành tập huấn luyện và kiểm tra
def train_test_split(X, y, test_size=0.3, random_state=42):
    np.random.seed(random_state)
    indices = np.random.permutation(X.shape[0])
    test_size = int(X.shape[0] * test_size)
    test_indices = indices[:test_size]
    train_indices = indices[test_size:]
    return X[train_indices], X[test_indices], y[train_indices], y[test_indices]

# Huấn luyện mô hình Linear Regression
def linear_regression(X, y, lambda_reg=1e-5):
    # Tính toán hệ số hồi quy (beta) sử dụng phương pháp bình phương tối thiểu với regularization
    X_transpose = X.T
    identity_matrix = np.eye(X.shape[1])
    beta = np.linalg.inv(X_transpose @ X + lambda_reg * identity_matrix) @ X_transpose @ y
    return beta

# Tạo tên các đặc trưng bậc cao (bao gồm bình phương và tương tác)
def get_feature_names(feature_columns, degree):
    from itertools import combinations_with_replacement
    feature_names = ["1"]
    for d in range(1, degree + 1):
        for items in combinations_with_replacement(feature_columns, d):
            feature_names.append("*".join(items))
    return feature_names

# Đánh giá mô hình
def mean_absolute_error(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def r2_score(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    return 1 - (ss_residual / ss_total)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# In kết quả đánh giá
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"R-squared: {r2:.2f}")




X_poly = polynomial_features(X_scaled, degree=4)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=42)

beta = linear_regression(X_train, y_train)

# In công thức hồi quy đầy đủ
print("\nCông thức hồi quy đầy đủ:")
intercept = beta[0]
coefficients = beta[1:]
# Dự đoán trên tập kiểm tra
y_pred = X_test @ beta
y_pred_3 =y_pred
feature_names_poly = get_feature_names(feature_columns, degree=4)

# In công thức với các đặc trưng bậc cao
formula = f"Sleep_Quality = {intercept:.2f} "
for coef, feature in zip(coefficients, feature_names_poly):
    if coef != 0:  # Bỏ qua các hệ số bằng 0
        formula += f"+ ({coef:.2f}) * {feature} "

print(formula)
```
